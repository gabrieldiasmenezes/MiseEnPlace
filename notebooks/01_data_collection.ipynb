<h2 style="color: #007acc;">2. Data Quality Diagnosis</h2>

Before any analysis, we need to understand the quality of the data.
The first check is identifying missing values — and deciding
whether they require treatment or carry analytical meaning.

```python
df.isnull().sum()
```

> **Findings:**
> - Only 3 columns contain null values
> - `PhoneNumber` and `WebsiteUrl` are irrelevant — will be dropped
> - `FacilitiesAndServices` (1,122 nulls) is relevant and requires
>   deeper investigation: are nulls distributed evenly across all
>   restaurants, or concentrated in specific award categories?
> 
> This question will be answered after we map the award categories.
> 
> ---

```python
order_award=['3 Stars','2 Stars','1 Star','Bib Gourmand','Selected Restaurants']
category_award=df['Award'].value_counts().reindex(order_award)
category_award
```

> **Findings:**
> - The dataset is heavily imbalanced: 11,507 unawarded restaurants
>   versus only 154 with 3 Stars
> - This imbalance means absolute counts alone can be misleading —
>   proportional analysis will be essential throughout this project
> - With award categories mapped, we can now investigate
>   the FacilitiesAndServices null distribution by category
> 
> ---

## 4. Columns Under Investigation

Three columns require deeper analysis before we decide
whether to keep or drop them: `Latitude/Longitude vs Address`,
`GreenStar`, and `Cuisine`.

---

### 4.1 Location: Address vs Latitude/Longitude

Both columns reference restaurant location, but serve different purposes.
`Latitude` and `Longitude` offer coordinate precision — essential for
map visualization in Power BI and for AI-based regional pattern analysis.
`Address` is human-readable but harder to process programmatically.

**Decision:** Keep `Latitude` and `Longitude`. Drop `Address`.

---

### 4.2 GreenStar

GreenStar is a separate Michelin award recognizing restaurants committed
to sustainable practices. The question is: does having a GreenStar
correlate with higher Michelin awards — making it relevant for
investor analysis?

```python
print("FREQUNECIA ABSOLUTA")
df['GreenStar'].value_counts()
```

```python
print("FREQUENCIA RELATIVA")
(df['GreenStar'].value_counts(normalize=True)*100).round(2)
```

Less than 5% of restaurants have a GreenStar.
At first glance this could suggest GreenStar belongs to awarded restaurants —
but that assumption needs verification. Let's check the distribution
across award categories.

```python
green_star_award=pd.crosstab(df['Award'],df['GreenStar']).reindex(order_award)
green_star_award
```

> **Finding:** No meaningful relationship between GreenStar and award level.
> The ratio of 0s to 1s is consistent across all categories —
> 3-Star restaurants actually have fewer GreenStars than Selected Restaurants.
> 
> **Decision:** `GreenStar` is irrelevant for this analysis. Will be dropped.
> 
> ---

```python
print(df['Cuisine'].head(20).to_string())
```

```python
print("\n=== Rows with multiple cuisines ===")
multiple=df['Cuisine'].str.contains(',',na=False).sum()
print(f"{multiple} rows contain more than one cuisine ({(multiple/len(df)*100).round(1)}%)")
```

> **Finding:** A significant portion of restaurants list multiple cuisines
> per row in inconsistent order
> (e.g. `Creative, Modern Cuisine` and `Modern Cuisine, Creative`
> represent the same combination).
> 
> Additionally, the column mixes culinary **styles** (Creative, Contemporary)
> with culinary **specialties** (French, Japanese, Italian)
> — making direct comparison unreliable and potentially misleading
> for investment decisions.
> 
> **Decision:** `Cuisine` will be dropped from the analysis.
